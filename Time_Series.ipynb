Simple Moving Average

```python
df_tesla['Open'].plot()
```

```python
df_tesla['Open:10 days rolling']=df_tesla['Open'].rolling(window=10, min_periods=1).mean()
```

```python
df_tesla[['Open:10 days rolling','Open']].plot()
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(window=30, min_periods=1).mean()
```

```python
df_tesla['Open:50 days rolling']=df_tesla['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df_tesla[['Open:10 days rolling','Open:30 days rolling','Open:50 days rolling','Open']].plot(figsize=(18,25))
```

Cummulative Moving Average- expanding()

```python
df_tesla['Open'].expanding().mean().plot()
```

Exponentail moving average

```python
df_tesla['EMA_0.1']=df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1']].plot()
```

Exponential weighted moving average

```python
df_tesla['EWMA_5 days']=df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open','EMA_0.1','EWMA_5 days']].plot()
```
